```python
qsmax = float(input("Enter qsmax in g/g*h :")) 
print(qsmax)
```

```python
Ks = float(input("Enter Ks in g/l :")) 
print(Ks)
```

```python
µmax = float(input("Enter µmax in 1/h :"))
print(µmax)
```

```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt



def model(y,t):
    Cs = y[0]
    Cx = y[1]
    qs = qsmax* Cs/(Cs+Ks)
    µ = µmax*Cs/(Cs+Ks)
    dCsdt = -qs*Cx
    dCxdt = µ*Cx
    return (dCsdt, dCxdt)
# bei return array([dCsdt, dCxdt]) steht da name 'array' is not defined, weil man from numpy import * machen muss vorher
# oder np.array

y0 = [50, 30]
t = np.linspace (0, 15, 100)
result = odeint(model, y0, t)
print (result)

Cs = result[:,0]
Cx = result[:,1]

plt.plot(t, Cs)
plt.plot(t, Cx)
plt.xlabel("time in h")
plt.ylabel("conc in g/l")
plt.legend(["Cs", "Cx"])
plt.title("Biomassconc and substrateconc over time")

plt.show
plt.close

#anstatt legend kann man auch bei plt.plot label= "CS"  oder "Cx" machen
# dann geht nocht plt.legend(loc= "best")
#plt.grid für gitter
# dann geht noch bei plt.lot(x,y, "r-"   oder "b--") für farbe und gestrichelt







```
